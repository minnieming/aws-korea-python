## 파이플롯 
맷플롯립에 속해있는 것

```python
x = range(100) # range함수 : 숫자를 나열하는 함수. 0-99까지 1씩 증가한다 (원하면 시작, 증가등을 설정할 수 있다)
y = range(100)
plt.plot(x,y)
```

```python
import numpy as np

x_1 = range(100)
y_1 = [np.cos(value) for value in x_1]

x_2 = range(100)
y_2 = [np.sin(value) for value in x_2]

plt.plot(x_1, y_1)
plt.plot(x_2, y_2)
plt.show()
```

그림(도화지. 바탕이라고 보면 될듯) figure (fig), 축 axis (ax)

```python
fig, ax = plt.subplots()

x_1 = range(100)
y_1 = [np.cos(value) for value in x]

ax.plot(x_1, y_1)
ax.set(title = 'cos graph', xlabel = 'x', ylabel='y')

plt.show()
```

```python
fig = plt.figure()
fig.set_size_inches(10,10)

ax_1 = fig.add_subplot(1,2,1) # (행,열,처음 들어가는 그림을 ax_1으로 할꺼다) 라는 의미
ax_2 = fig.add_subplot(1,2,2)

ax_1.plot(x_1, y_1, c='b')
ax_2.plot(x_2, y_2, c="g")

ax_1.plot(x_1, y_1, c='b') # 그래프를 어떻게 만들지 정하는 것
ax_2.plot(x_2, y_2, c='g')
plt.show()
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2)
print(ax)
```

```python
x = np.linspace(-1,1,100) # (여기부터, 여기까지, 100개씩 나눠서 하라는 뜻)
y_1 = np.sin(x)
y_2 = np.cos(x)
y_3 = np.tan(x)
y_4 = np.exp(x)

fig, ax = plt.subplots(2,2)

ax[0,0].plot(x, y_1)
ax[0,1].plot(x, y_2)
ax[1,0].plot(x, y_3)
ax[1,1].plot(x, y_4)
```

```python
ax1 = plt.subplot(3,2,1)
plt.plot(x, y_1)

ax2 = plt.subplot(3,2,2)
plt.plot(x, y_1)

ax3 = plt.subplot(3,1,2)
plt.plot(x, y_1)

ax4 = plt.subplot(3,2,5)
plt.plot(x, y_1)

ax5 = plt.subplot(3,2,6)
plt.plot(x, y_1)

plt.show()
```

```python
x_1 = range(100)
y_1 = [value for value in x_1]

x_2 = range(100)
y_2 = [value + 100 for value in x_2]

plt.plot(x_1, y_1, color = "#82cbe7")
plt.plot(x_2, y_2, c='c')

plt.show()
```

```python
# 선의 형태
plt.plot(x_1, y_1, c = "b", linestyle = 'dashed')
plt.plot(x_2, y_2, c='c', ls = 'dotted')

plt.show()
```

```python
# 선의 형태
plt.plot(x_1, y_1, c = "b", linestyle = 'dashed')
plt.plot(x_2, y_2, c='c', ls = 'dotted')

plt.title('Two lines', fontsize = 20)
plt.show()
```

```python
from matplotlib import font_manager, rc
font_path = 'c:/Windows/Fonts/malgun.ttf'
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)
plt.rcParams['axes.unicode_minus'] = False # 한글로하면 마이너스가 깨지는데, 그거 방지하는것. -> 한글 설정 다음에 넣어야 한다 

fig = plt.figure()
fig.set_size_inches(10,10)

ax_1 = fig.add_subplot(1,2,1)
ax_2 = fig.add_subplot(1,2,2)

ax_1.plot(x_1, y_1, c='b')
ax_1.set_title('그림 1')
ax_2.plot(x_2, y_2, c='g')
ax_2.set_title('figure 2')


plt.show()
```

```python
plt.plot(x_1, y_1, color = 'b', linestyle = 'dashed', label = 'line_1')
plt.plot(x_2, y_2, color='r', linestyle = 'dotted', label = 'line_2')

plt.legend(loc=('best')) # legend == 범례 지정

plt.title('$y = ax+b$') #$로 감싸면 수학식이 만들어진다.
plt.xlabel('$x_{line}$')
plt.ylabel('y_line')
```

```python
# 산점도 그래프 -> 분포도 그릴때 많이 쓴다. 
data_1 = np.random.rand(512,2)
data_2 = np.random.rand(512,2)

print(data_1.shape, data_1) # shape : 모양을 보여준다. 
print(data_2.shape, data_2)
```

```python
plt.scatter(data_1[:,0], data_1[:,1], c='b', marker='x') # data_1[:(전체에서), 해당 리스트의 0번째 있는것]을 이 그래프의 x축에, 그 그래프를 y축에 넣기
plt.scatter(data_2[:,0], data_2[:,1], c='r', marker='o')

plt.show()
```

```python
# 산접도 그래프를 응용해 버블 그래프를 그릴 수 있다. 
# x,y축 뿐만 아니라 해당 포인트의 넓이를 정할 수 있다. 

n = 50
x = np.random.rand(n)
y = np.random.rand(n)
colors = np.random.rand(n)
area = np.pi*(15*np.random.rand(n))**2

plt.scatter(x,y,s=area, c= colors, alpha=0.5)

plt.show()
```

```python
# 막대그래프

data  = [[5., 25.,50.,20,], [4.,23.,51,17], [6.,22.,52.,19]] # 5. == 5.0을 의미하는 것

x=np.arange(0,8,2)

plt.bar(x + 0.00, data[0], color = 'b', width = 0.50)
plt.bar(x + 0.50, data[1], color = 'g', width = 0.50)
plt.bar(x + 1.0, data[2], color = 'r', width = 0.50)

plt.xticks (x+0.50, ('a', 'b', 'c', 'd'))

plt.show()
```

```python
data = np.array ([[5., 25.,50.,20,], [4.,23.,51,17], [6.,22.,52.,19]])

color_list = ['b', 'g', 'r']
data_label = ['A', 'B', 'C']
x = np.arange(data.shape[1])

for i in range(3):
    plt.bar(x, data[i], bottom = np.sum(data[:i], axis = 0), color = color_list[i], label = data_label[i])
    print(np.sum(data[:i], axis = 0))

plt.legend()
plt.show()
```

```python
# 원 그래프

ratio = [34,32,1,18]
labels = ['apple', 'banana', 'melon', 'grapes']
explode = [0,0.1,0,0]
colors = ['silver', 'gold', 'whitesmoke', 'lightgray']

plt.pie(ratio, labels=labels, autopct='%.1f%%', startangle = 260, counterclock=False, explode=explode, shadow=True, colors = colors)

plt.show()
```
