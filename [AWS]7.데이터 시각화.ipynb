#### 1.1 파이플롯
- 맷플롯립을 이용할 때 가장 기본이 되는 구조는 파이플롯(pyplot)이다.

````python
import matplotlib.pyplot as plt # matplotlib 모듈 호출

X = range(100)
Y = range(100)
plt.plot(X, Y)
````

````python
import numpy as np # numpy 모듈 호출

X_1 = range(100)
Y_1 = [np.cos(value) for value in X_1]

X_2 = range(100)
Y_2 = [np.sin(value) for value in X_2]

plt.plot(X_1, Y_1)
plt.plot(X_2, Y_2)
plt.show()
````

#### 1.2 그림과 축
- 그림(figure)는 그래프를 작성하는 밑바탕이 되는 툴이고, 축(axis)은 일종의 캔버스로 실제 그래프를 작성하는 공간이다.
- subplots함수를 이용해 하나의 figure에 여러개의 그래프를 그릴 수  있다.

````python
fig, ax = plt.subplots() # (1) figure와 axes 객체 할당

X_1 = range(100)
Y_1 = [np.cos(value) for value in X]

ax.plot(X_1, Y_1) # (2) plot 함수를 사용하여 그래프 생성
ax.set(title='cos graph', # (3) 그래프 제목, X축 라벨, Y축 라벨 설정
       xlabel='X',
       ylabel='Y');

plt.show() # (4) 그래프 출력
````

image.png

````python
fig = plt.figure() # (1) figure 반환
fig.set_size_inches(10,10) # (2) figure의 크기 지정

ax_1 = fig.add_subplot(1,2,1) # (3) 첫 번째 그래프 생성
ax_2 = fig.add_subplot(1,2,2) # (4) 두 번째 그래프 생성

ax_1.plot(X_1, Y_1, c="b") # (5) 첫 번째 그래프 설정
ax_2.plot(X_2, Y_2, c="g") # (6) 두 번째 그래프 설정
plt.show() # (7) 그래프 출력
````

````python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-1,1,100) # (1) x 값과 y_n 값 생성
y_1 = np.sin(x)
y_2 = np.cos(x)
y_3 = np.tan(x)
y_4 = np.exp(x)

fig, ax = plt.subplots(2, 2) # (2) 2×2 figure 객체를 생성

ax[0, 0].plot(x, y_1) # (3) 첫 번째 그래프 생성
ax[0, 1].plot(x, y_2) # (4) 두 번째 그래프 생성
ax[1, 0].plot(x, y_3) # (5) 세 번째 그래프 생성
ax[1, 1].plot(x, y_4) # (6) 네 번째 그래프 생성
````

````python
ax1 = plt.subplot(3,2,1) # (1) 첫 번째 공간에 axes 생성
plt.plot(x, y_1)
ax2 = plt.subplot(3,2,2) # (2) 두 번째 공간에 axes 생성
plt.plot(x, y_1)
ax3 = plt.subplot(3,1,2) # (3) 두 번째 공간에 axes 생성
plt.plot(x, y_1)
ax4 = plt.subplot(3,2,5) # (4) 다섯 번째 공간에 axes 생성
plt.plot(x, y_1)
ax5 = plt.subplot(3,2,6) # (5) 여섯 번째 공간에 axes 생성
plt.plot(x, y_1)

plt.show()
````

### 2 맷플롯립으로 그래프 그리기
- 맷플롯립은 다양한 형태의 그래프 커스터마이징이 가능하다. 기본적으로 색상이나 그래프의 모양을 바꾸는 방법들을 학습한다.

#### 2.1 색상
- color 또는 c 매개변수를 사용해서 색상을 변경할 수 있다.
- 16진수의 RGB값을 입력하는 방법
- 이미 설정된 약어를 사용하는 방법

````python
X_1 = range(100)
Y_1 = [value for value in X_1]

X_2 = range(100)
Y_2 = [value + 100 for value in X_2]

plt.plot(X_1, Y_1, color="#75C758")
plt.plot(X_2, Y_2, c="c")

plt.show()
````

#### 2.2 선의 형태
- linestyle 또는 ls를 이용한 매개변수 입력

````python
plt.plot(X_1, Y_1, c="b",
         linestyle="dashed")
plt.plot(X_2, Y_2, c="r",
         ls="dotted")

plt.show()
````

#### 2.3 제목

````python
plt.plot(X_1, Y_1, color="b",
         linestyle="dashed")
plt.plot(X_2, Y_2, color="r",
         linestyle="dotted")

plt.title("Two lines")
plt.show()
````

````python
# matplotlib 한글 폰트 오류 문제 해결
from matplotlib import font_manager, rc
import matplotlib.font_manager as fm
font_path = "c:/Windows/Fonts/malgun.ttf"   #폰트파일의 위치
font_name = font_manager.FontProperties(fname=font_path).get_name()

rc('font', family=font_name)

fig = plt.figure()
fig.set_size_inches(10,10)

ax_1 = fig.add_subplot(1,2,1)
ax_2 = fig.add_subplot(1,2,2)

ax_1.plot(X_1, Y_1, c="b")
ax_1.set_title("그림 1")
ax_2.plot(X_2, Y_2, c="g")
ax_2.set_title("Figure 2")

plt.show()
````

#### 2.4 범례
- 그래프에 대한 설명

````python
plt.plot(X_1, Y_1,
         color="b",
         linestyle="dashed",
         label='line_1')
plt.plot(X_2, Y_2,
         color="r",
         linestyle="dotted",
         label='line_2')
plt.legend(
    shadow=True,
    fancybox=False,
    loc="upper right")

plt.title('$y = ax+b$')
plt.xlabel('$x_{line}$')
plt.ylabel('y_line')
````

#### 3.2 막대그래프

#### 3.3 누적 막대그래프

````python
data = np.array([[5., 25., 50., 20.],
                 [4., 23., 51., 17],
                 [6., 22., 52., 19]])

color_list = ['b', 'g', 'r']
data_label = ["A","B","C"]
X = np.arange(data.shape[1])

for i in range(3):
    plt.bar(X, data[i],
            bottom = np.sum(data[:i], axis=0), # bottom은 data[i]의 값이 시작되는 위치
            color = color_list[i],
            label=data_label[i])
    print(np.sum(data[:i], axis=0))
    
plt.legend()
plt.show()
````

## 02 시본
- 맷플롯립과 동일하게 시각화를 할 수 있게 해주는 도구 맷플롭립과의 차이점은 좀 더 보기 좋은 그래프를 지원한다.

### 1 시본의 기본
- 시본은 맷플롯립을 바탕으로 다양한 함수를 사용할 수 있는 일종의 래퍼(wrapper)모듈이다.   
- seaborn 호출
```python
import seaborn as sns
```

````python
# import numpy as np # numpy 모듈 호출
# import pandas as pd # pandas 모듈 호출
# import matplotlib.pyplot as plt # matplotlib 모듈 호출
import seaborn as sns # (1) seaborn 모듈 호출

fmri = sns.load_dataset("fmri") # (2) fmri 데이터셋 사용(fMRI 데이터셋은 기능적 자기공명영상(functional Magnetic Resonance Imaging, fMRI) 기술을 통해 얻어진 데이터)
sns.set_style("whitegrid") # (3) 기본 스타일 적용
sns.lineplot(x="timepoint", y="signal", data=fmri) # (4) 선그래프 작성
````

image.png

- 시본은 맷플롯립의 figure 객체나 axes 객체와 같은 개념이 없다.
- x, y 축 라벨이 자동 생성된다.


````python
fmri.sample(n=10, random_state=1)
````

````python
sns.lineplot(x="timepoint", y="signal", hue="event", data=fmri)
````

- 이럴 때 맷플롯립을 사용하면 상당히 복잡하게 표현해야 하지만 시본은 단순히 hue라는 매개변수만 추가하면 그래프를 그릴 수 있다.

### 2 시본에서 사용하는 그래프
- 시본에서 사용할 수 있는 회귀 그래프, 산점도, 비교 그래프, 막대그래프 등의 다양한 그래프에 대해 알아본다.

#### 2.1 회귀 그래프
- 단순한 선그래프(lineplot)외에도 시본에는 미리 정의된 다양한 그래프를 사용할 수 있다.
- 그중 회귀 그래프(regression plot)는 회귀식을 적용하여 선형추세선을 그래프에 함께 작성하는 것이다. 함수로 regplot을 사용.

````python
tips = sns.load_dataset("tips")
sns.regplot(x="total_bill", y="tip", data=tips, x_ci=95)
# x_ci -> 신뢰구간 비율을 나타내는 매개변수
````

### 3 사전 정의된 그래프
- 사전 정의된 그래프는 맷플롯립 관점에서 하나 이상의 그래프들을 합쳐서 다양한 정보를 줄 수 있는 그래프를 의미.  
- 범주형 데이터에서 매우 유용하게 쓸 수 잇는 그래프들이 많다.

#### 3.1 분포를 나타내는 그래프: 바이올린 플롯과 스웜 플롯
- 바이올린 플롯(violin plot)은 데이터의 분포를 바이올린과 비슷한 형태로 보여주는 그래프로, 상자그림과 분포도를 한 번에 나타낼 수 있다.
- violinplot함수를 사용

````python
sns.violinplot(x="day", y="total_bill", hue="smoker",
               data=tips, palette="muted")
````

- 같은 형태로 스웜 플롯(swarm plot)이 있다.
- 스웜 플롯은 바이올린 플롯과 같은 형태지마, 산점도를 이용해서 데이터의 분포를 나타내는 차이점이 있다.
- 스웜 플롯은 매개변수 hue와 함께 사용하여 두 개 이상의 범주형 데이터를 비교할 때, 데이터의 형태를 점이 겹치지 않게 정리하여 각 데이터의 영역별로 얼마나 많은 데이터가 있는지 직관적으로 차악 할 수 있다.
- swarmplot 함수를 사용

````python
sns.swarmplot(x="day", y="total_bill", 
              hue="smoker", data=tips, palette="muted")
````

#### 3.2 다양한 범주형 데이터를 나타내는 페싯그리드
- 맷플롯립과 달리 그래프의 틀만 제공하여 적당한 그래프를 그릴 수 있다. 이를 위해 패싯그리드라고 부르는 FacetGrid함수를 사용할 수 있다.
- sex, smoker, day, time등은 범주형 데이터   
image.png

````python
tips.head()
````

- 다음 데이터는 tips라는 객체에 저장되어 있다. 기본적인 데이터 표현 틀을 만들기 위해 다음과 같이 할 수 있다. 
- 기본적으로 col과 row에 범주형 데이터를 너르면 해당 열의 데이터 종류만큼 'm x n'의 그래프 틀이 생긴다.

````python
g = sns.FacetGrid(tips, col="time",
                  row="sex")
````

- 그리드 생성 후 실제 그래프를 만들 때는 일종의 맵(map) 개념을 사용.
즉, 생성된 그래프를 각 FacetGrid에 있는 개별 영역에 넣는 구조로 되어 있음.

````python
g = sns.FacetGrid(tips,
                  col="time",
                  row="sex")

g.map(sns.scatterplot,
      "total_bill", "tip")
````

- histplot로 FacetGrid 그리기

````python
g = sns.FacetGrid(tips, col="time",
                  row="sex")

g.map_dataframe(sns.histplot,
                x="total_bill")
````

- 문법은 맷플롯립이나 시본과 유사하다.

````python
import plotly.express as px
df = px.data.iris()                 # iris is a pandas DataFrame
fig = px.scatter(df, x="sepal_width", y="sepal_length")
fig.show()
````

- plotly.express를 호출 express는 사실상 시본과 같이 그래프를 호출 할 수 있는 래퍼 모듈

````python
fig = px.parallel_coordinates(df, color="species_id", labels={"species_id": "Species",
                "sepal_width": "Sepal Width", "sepal_length": "Sepal Length",
                "petal_width": "Petal Width", "petal_length": "Petal Length", },
                             color_continuous_scale=px.colors.diverging.Tealrose,
                             color_continuous_midpoint=2)
fig.show()
````
