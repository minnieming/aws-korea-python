```python
import pandas as pd
```

```python
df = pd.read_excel('C:/Users/Admin/AWS Korea_sesac/전출지_전입지_시도_별_이동자수_1970_2024.xlsx')
df.head(2) # 이렇게 하면 위에서 두줄이 된다 
# 이 형식을 PANDAS에서는 데이터 형식이라고 한다. 
```

```python
df = df.iloc[1:]
df.head(2)
# NaN : null값이라는 뜻 (셀병합이 되어 있는 경우가 그럴 수가 있다.) -> 이런경우에 pandas로 채워주는데 그때 쓰는게 ffill (밑에 나온다)
```

```python
df.info() # 형태와 자료형을 보여준다
# 17 non-null : null이 아닌게 18개다 라는 뜻
```

```python
df = df.ffill() # 이렇게 되면 '전국'으로 채워진다
df.head()
```

```python
# '-' 이게 있으면 에러가 날 수 있어서 바꿔주는게 좋다. (null 값은 없는게 낫다) -> 이걸 0으로 바꾸는게 낫기때문에 바꾸기
df.replace('-', 0, inplace=True)
df.head()
```

```python
df.info() # 자료형이 object -> int로 바꼈다
```

```python
# 필터링 기능
df_jeju = df[(df['전출지별'] == '제주특별자치도') & (df['전입지별']!='제주특별자치도')]

df_jeju.head()
```

```python
# 삭제하기
del df_jeju['전출지별']
df_jeju.head()
```

```python
df_jeju = df_jeju.set_index('전입지별', drop = True) # 해당 인덱스의 drop을 ok 한다는 뜻
df_jeju.head()
```

```python
df_jeju = df_jeju.drop('전국', axis=0) # axis : 행(0)인지 열(1)인지 정하는것
df_jeju.head()
```

```python
df_jeju_top4 = df_jeju.sort_values(by='2024', ascending=False).head(4)
df_jeju_top4
```

```python
df_jeju_top4.T.plot(kind='area', stacked=False, figsize=(20,10), alpha=0.2)
plt.title('제주특별자치도 타지역 이동 인구 TOP4', size = 30)
plt.xlabel('기간')
plt.ylabel('이동자수')
plt.legend(loc='best', fontsize=15)
plt.show()
```
