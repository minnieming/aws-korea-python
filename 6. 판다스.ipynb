# 3. 그룹별 집계

group by : 데이터로부터 동일한 객체를 가진 데이터만 뽑아 데이터 추출하는 기능

```python
ipl_data = {'team':['riders','riders', 'devils','devils','kings','kings','kings','kings','riders','royals','royals','royals'],
            'rank':[1,2,2,3,3,4,1,1,2,4,1,2],
            'year':[2014,2015,2014,2015,2014,2015,2016,2017,2016,2014,2015,2017],
            'points':[876,789,863,673,741,812,756,788,694,701,804,690]}

df = pd.DataFrame(ipl_data)
df
```

```python
# 각 팀마다 포인트를 얼마나 얻었는지 더해주는 것.
# 기준열 / 적용받는열 / 어떻게 적용할껀지

df.groupby('team')['points'].sum()
```

멀티 인덱스 (인덱스가 하나 이상인것) 그룹별 집계

```python
multi_groupby = df.groupby(['team', 'year'])['points'].sum()
multi_groupby
```

```python
multi_groupby.index
```

```python
# 비어있는 값에 nan값이 대체된다.
multi_groupby.unstack()
```

```python
# 인덱스간의 레벨 변겨
multi_groupby.swaplevel().sort_index()
```

```python
multi_groupby.groupby(level=1).sum()
```
